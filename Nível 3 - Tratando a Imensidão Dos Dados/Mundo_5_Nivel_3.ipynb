```python
import pandas as pd

dados = pd.read_csv('Pacers.csv', sep=';', engine='python', encoding='utf-8')
dados['Calories'].fillna(0, inplace=True)
dados['Date'].fillna('1900/01/01', inplace=True)

dados['Date'] = dados['Date'].replace('20201226', "'2020/12/26'")

dados['Date'] = pd.to_datetime(dados['Date'], errors='coerce')

dados.dropna(subset=['Date'], inplace=True)

print(dados)
```
